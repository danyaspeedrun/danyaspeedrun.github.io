```python
import requests as r
import itertools
import os
import json
from glob import glob
```

Fetch the games

```python
years = [2020]
months = [9,10,11]
for (yyyy,mm) in itertools.product(years,months):
    remote_base_url = 'https://api.chess.com/pub/player/senseidanya/'
    game_url = 'games/{:04}/{:02}'.format(yyyy,mm)
    print(remote_base_url + game_url)
    
    xx = r.get(remote_base_url + game_url)
    print(xx.status_code)
    if xx.status_code == 200:
        os.makedirs(os.path.dirname(game_url), exist_ok=True)
        with open(game_url + '.json','wb') as f:
            f.write(xx.content)    
```

```python
tmp = {}
tmp['id'] = []
tmp['vod_start_posix'] = []
tmp['vod_end_posix'] = []
tmp['vod_thumb'] = []
tmp['vod_url'] = []
for fname in glob('./vods/*.json'):
    with open(fname) as f:
        vod_json = json.load(f)
        tmp['id'].append(vod_json['id'][1:])
        tmp['vod_start_posix'].append(vod_json['timestamp'])
        tmp['vod_end_posix'].append(vod_json['timestamp']+vod_json['duration'])
        tmp['vod_thumb'].append(vod_json['thumbnails'][-1]['url'])
        tmp['vod_url'].append('https://www.twitch.tv/videos/' + tmp['id'][-1])
        
vod_df = pd.DataFrame(tmp)

vod_df['vod_start'] = pd.to_datetime(vod_df['vod_start_posix'], unit='s')
vod_df['vod_end'] = pd.to_datetime(vod_df['vod_end_posix'], unit='s')
vod_df = vod_df.sort_values('vod_start')
vod_df
```

```python
joined_df = pd.merge_asof(games_df,vod_df, direction='backward', left_on='dt', right_on='vod_start')
joined_df['vod_offset'] = (joined_df['dt']-joined_df['vod_start']).dt.total_seconds().astype('int')
joined_df['vod_link_with_offset'] = joined_df['vod_url'] + "?t=" + joined_df['vod_offset'].astype('str') + "s"
joined_df
```

```python
header_str = """
<!doctype html>

<html lang="en">
<head>
  <meta charset="utf-8">
  
  <title>The HTML5 Herald</title>
  <meta name="description" content="The HTML5 Herald">
  <meta name="author" content="SitePoint">
  
  <style>
table {
  font-family: arial, sans-serif;
  border-collapse: collapse;
  width: 100%;
}

td, th {
  border: 1px solid #dddddd;
  text-align: left;
  padding: 8px;
  text-align: center;
}

tr:nth-child(even) {
  background-color: #aaaaaa;
}



</style>
  
  
  
</head>

<body>
  
  <p> Hello World! 552</p>
  
   <table style="width:100%">
  <tr>
    <th>Game #</th>
    <th>Date</th>
    <th>White</th>
    <th>Black</th>
    <th>Result</th>
    <th>Opening</th>
    <th>Vod</th>
  </tr>
"""

footer_str = """  
</table> 
</body>
</html>
"""

#https://www.w3schools.com/html/html_tables.asp

html_str = header_str 

for idx,row in joined_df.iterrows():
    html_str += "<tr>"
    html_str += "<td>{}</td>".format(idx)
    html_str += "<td>{}</td>".format(row['UTCDate_str'])
    html_str += "<td>{maybe_strong}<p>{white_name}</p><p>{WhiteElo}</p></td>".format(
        maybe_strong = "<strong>" if row['result'] == "1-0" else "",
        white_name = row['white'],
        WhiteElo = row['WhiteElo']
    )
    html_str += "<td>{maybe_strong}<p>{black_name}</p><p>{BlackElo}</p></td>".format(
        maybe_strong="<strong>" if row['result'] == "0-1" else "",
        black_name=row['black'],
        BlackElo=row['BlackElo']
    )    
    html_str += "<td>{}</td>".format(row['result'])
    html_str += '<td><a href={}  target="_blank" ><img src={}><p>{}</p></a></td>'.format(row['Op'],row['opening_img'],row['opening_name'])
    html_str += '<td><a href={}  target="_blank" ><img src={}></a></td>'.format(row['vod_link_with_offset'],row['vod_thumb'])
    html_str += "</tr>"
    
    
html_str += footer_str

with open("out.html",'w') as f:
    
    f.write(html_str)
```
